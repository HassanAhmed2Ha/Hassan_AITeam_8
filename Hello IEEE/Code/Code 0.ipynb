```python
list_0 = [

    [1,2,1],
    [3,4,5],
    [6,7,8],
    ]

list_1 = [
    [2,3,4],
    [6,9,7],
    [9,6,4],
         
         ]
```

```python
def matsum(arr1, arr2):
    result = [[arr1[i][j] +arr2[i][j] for j in range(len(arr1[0]))] for i in range(len(arr1))]
    return result
print(matsum(list_0, list_1))
```

```python
def matsub(arr1, arr2):
    result = [[arr1[i][j] - arr2[i][j] for j in range(len(arr1[0]))] for i in range(len(arr1))]
    return result
print(matsum(list_0, list_1))
```

```python
def scalarsum(scalar, arr):
    result = [[arr[i][j] + scalar for j in range(len(arr[0]))] for i in range(len(arr))]
    return result

print(scalarsum(10, list_0))
```

```python
def (scalar, arr):
    result = [[arr[i][j] - scalar for j in range(len(arr[0]))] for i in range(len(arr))]
    return result

print(scalarsum(10, list_0))
```

```python
def matmul(arr1, arr2):
    result = [[0 for j in range(len(arr2[0]))] for i in range(len(arr1))]
    for i in range(len(arr1)):
        for j in range(len(arr2[0])):
            for k in range(len(arr2)):
                result[i][j] = result[i][j] + (arr1[i][k] * arr2[k][j])
                
    return result

print(matmul(list_0, list_1))
```
