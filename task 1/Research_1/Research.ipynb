### Research Question 1 Why is Numpy so fast

When I researched this I found out that numpy is not just a library it is a high performance engine written in C and C++ 

Python loops are slow for big data because they are interpreted line by line and every single number is treated as a heavy object with type checking This is like carrying one brick at a time to build a house 

Numpy uses Vectorization which is like using a massive crane to move thousands of bricks at once It uses SIMD Single Instruction Multiple Data to process whole arrays in the CPUs hardware Also numpy stores data in contiguous memory blocks making it lightning fast for the processor to access everything in one go

## Research Question 2 The Multicollinearity Nightmare

I also looked into what happens when features are too similar like having house size in square meters and square feet in the same model This is called multicollinearity and it creates a huge mathematical problem for our Normal Equation

The Problem
In our formula we need to calculate the inverse of XT X If features are highly correlated this matrix becomes singular or non invertible in simple terms the math breaks because we are essentially trying to divide by zero 

The Solutions
Instead of just deleting data I found two professional ways to fix this
1 The Pseudo inverse Using np linalg pinv instead of inv This smart algorithm can find an approximate solution even when the matrix is singular
2 Regularization By adding a tiny bit of information a penalty term called lambda to the diagonal of the matrix we nudge it away from being singular and make it invertible again This is the foundation of Ridge Regression
