Alright now that we can see what we are dealing with lets get down to business First I need to convert this normal python list into a numpy array because numpy is basically magic and makes math super fast

```python
grades_array = np.array(grades)
```

```python
np.shape(grades)
```

Next up finding the averages Numpy makes this ridiculously easy using the axis parameter Axis 1 gives me the average for each student walking horizontally and axis 0 gives me the average for each subject walking vertically

```python
mean_stu=np.mean(grades,axis=1)
print(mean_stu)
```

```python
mean_sub=np.mean(grades,axis=0)
print(mean_sub)
```

Now here is a really cool trick called boolean masking Instead of writing long loops to find the top students I just tell numpy to give me the rows where the students average is strictly greater than 85

```python
top_stu = grades_array[mean_stu > 85]
print(top_stu)
```

Time for a combo move Im going to give everyone a 5 mark bonus using broadcasting which is literally just adding 5 to the whole array Then ill normalize the grades using the min max formula so everything is between 0 and 1 Finally ill flatten the whole matrix into a single line because why not

```python
grades_bonus = grades_array + 5
print(grades_bonus)
```

```python
grades_normalized = (grades_array - np.min(grades_array)) / (np.max(grades_array) - np.min(grades_array))
print(grades_normalized)
```

```python
grades_flat = grades_array.flatten()
print(grades_flat)
```

To wrap up this part I wanted to make the final numbers look good so I used seaborn to draw a heatmap Darker blue means higher grades simple and clean

```python


plt.figure(figsize=(8, 5))
sns.heatmap(grades_array, annot=True, cmap="Blues", fmt="g", cbar=True)
plt.title("Students Grades Heatmap")
plt.xlabel("Subjects (Columns)")
plt.ylabel("Students (Rows)")
plt.show()
```
