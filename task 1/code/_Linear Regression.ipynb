Predicting House Prices with Math

After finishing the grades analysis I decided to try something even more challenging predicting house prices based on house size This sounds like a job for machine learning but I wanted to understand the logic behind it first by using the Normal Equation

The plan is simple Convert my data to numpy arrays add a bias term to account for the intercept and then apply this beautiful matrix formula

$$\theta = (X^T X)^{-1} X^T y$$

I have the sizes 50 60 80 100 and 120 square meters and the prices 150 180 240 300 and 330 thousands Lets see if I can find the pattern

```python
import numpy as np
```

```python
# Preparing the house size (X) and price (y) data
X_raw = np.array([50, 60, 80, 100, 120])
y = np.array([150, 180, 240, 300, 330])

# Reshaping X into a column vector to make it a proper matrix
X = X_raw.reshape(-1, 1)

# Adding a column of ones to X for the bias (intercept) term
m = len(X)
X_b = np.append(np.ones((m, 1)), X, axis=1)

print("X matrix with bias term:\n", X_b)
```

The Normal Equation Step by Step

Now comes the magic part I am going to use numpy linear algebra functions to solve the equation step by step No black box libraries just pure matrix multiplication and inversion to find the theta parameters

```python
theta = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y

print("Computed Theta parameters:", theta)

# Now lets predict the price for a 90 square meter house
# We must include the 1 for the bias term in our input
X_new = np.array([1, 90])
predicted_price = X_new @ theta

print(f"The predicted price for a 90 sqm house is: {predicted_price:.2f} thousand")
```
