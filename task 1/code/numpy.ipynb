```python
import numpy as np
```

```python
grades =[[85, 78, 92, 88],
[70, 76, 80, 65],
[90, 88, 94, 91],
[60, 65, 58, 62],
[100, 95, 98, 97]] 
```

```python
grades_array = np.array(grades)
```

```python
np.shape(grades)
```

### احسب متوسط درجات الطلاب امشي بشكل افقي

```python
mean_stu=np.mean(grades,axis=1)
print(mean_stu)
```

### احسب متوسط درجات الطلاب امشي بشكل رأسي

```python
mean_sub=np.mean(grades,axis=0)
print(mean_sub)
```

```python
top_stu = grades_array[mean_stu > 85]
print(top_stu)
```
